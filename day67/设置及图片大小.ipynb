```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(20,8),dpi=80)
#figure图像图标的意思,这里指的就是我们画的图
#通过实例化一个figure并且传递参数,就能够在后台自动使用该figure实例
#在图形模糊的时候可以传入dpi参数,让图片更加清晰
x=range(2,26,2)
    #数据在x轴的位置,是一个可迭代对象
y=[15,13,14.5,17,20,25,26,26,24,22,18,15]
    #数据在y轴的位置
    #

#设置x的刻度
plt.xticks(range(2,26))
#设置y的刻度
plt.yticks(range(min(y),max(y)))
plt.plot(x,y)#传入xy的数值,通过plot绘制出折线图
# plt.savefig("./sig_size.png")#保存图片     存为svg这种图片格式,放大不会有锯齿
plt.show()
```
